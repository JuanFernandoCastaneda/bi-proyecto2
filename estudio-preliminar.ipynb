```python
import pandas as pd
import math
```

```python
data = pd.read_csv('./data/data_2010_2017.csv')
```

```python
data.head(30)
```

```python
capitalize = ['evento', 'edad', 'sexo']
data[capitalize] = data[capitalize].applymap(lambda x: str(x).capitalize())
data.head()
```

```python
title = ['codigo_depto', 'nombre_depto', 'codigo_muni', 'nombre_muni', 'codigo_evento']
data[title] = data[title].applymap(lambda x: str(x).title())
data.head()
```

```python
data.sexo.unique()
```

```python
# WTf????????????????????
sexosErroneos = ['Menor 1 año', 'De 1-4 años', 'De 5-14 años', 'De 15-44 años', 'De 45-64 años', 'De 65 y más', 'Edad desconocida']
data.sexo.replace(sexosErroneos, 'Indeterminado', inplace=True)
data.sexo.unique()
```

```python
data.codigo_depto.unique()
```

```python
data.nombre_depto.unique()
```

```python
data[(((data.codigo_depto == '05') | (data.codigo_depto == 5)) & (data.nombre_muni == 'Medellín') & (data.evento == 'Total') & (data.edad == 'Total') & (data.sexo == 'Total'))]
```

De aquí identificamos que hay dos valores válidos para el dpto de Antioquia según el año y del dpto 76.
Además de eso, vamos a convertir los 'Total' en los códigos en string válidos y extranjeros en 0.

```python
data['codigo_depto'].replace('Total', '1', inplace = True)
data[data['codigo_depto'] == 'Total']['codigo_depto']
```

```python
data['codigo_depto'].replace('Extranjeros', '75', inplace = True)
data[data['codigo_depto'] == 'Extranjeros']['codigo_depto']
```

```python
data['codigo_muni'].replace('Total', '000', inplace = True)
data['codigo_muni'] = data['codigo_muni'].str.slice(-3)
data['codigo_muni'] = data['codigo_depto'].astype(str) + data['codigo_muni']
data[data.codigo_muni == '05000'].head()
```

```python
data['codigo_evento'].replace('Total','0', inplace = True)
data[data['codigo_evento'] == 'Total']['codigo_evento']
```

```python
data['codigo_depto'] = data['codigo_depto'].astype(int)
data['codigo_muni'] = data['codigo_muni'].astype(int)
data['codigo_evento'] = data['codigo_evento'].astype(int)
```

```python
data[((data.nombre_depto == 'Total') & (data.evento == 'Total') & (data.edad == 'Total') & (data.sexo == 'Total'))]
```

De aquí podemos concluir que hay casos que están divididos porque sí, así que toca sumarlos.

```python
columnasGroup = data.columns.drop('num_casos').to_list()
data = data.groupby(by=columnasGroup).sum().reset_index().sort_values(by=columnasGroup)
data.head(10)
```

```python
# Corremos la misma query y comprobamos que se redujeron los repetidos
data[((data.nombre_depto == 'Total') & (data.evento == 'Total') & (data.edad == 'Total') & (data.sexo == 'Total'))].sort_values(by=['anio'])
```
